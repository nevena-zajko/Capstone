```python
import pandas as pd
import datetime
```

```python
df_holidays_0 = pd.read_csv('data/holiday_data/holidays.csv')
df_holidays_0.head()
```

```python
df_holidays_0['Date'] =  pd.to_datetime(df_holidays_0['Date'], infer_datetime_format=True)
df_holidays_0['Date'] =  df_holidays_0['Date'].dt.date
df_holidays_0.drop_duplicates(inplace=True)
df_holidays_0['TimeOfDay'] = 0
df_holidays_0.head()
```

```python
# duplicate for each time of day
df_holidays_1 = df_holidays_0.copy()
df_holidays_1['TimeOfDay'] = 1
df_holidays_2 = df_holidays_0.copy()
df_holidays_2['TimeOfDay'] = 0
df_holidays_3 = df_holidays_0.copy()
df_holidays_3['TimeOfDay'] = 3
```

```python
# we want to consider the first 6 hrs after a holiday as a holiday as well
df_next_day_0 = df_holidays_0.copy()
df_next_day_0['Date']=df_next_day_0.Date + datetime.timedelta(days=1)
df_next_day_0['TimeOfDay'] = 0
```

```python
#join dataframes
df = pd.concat([df_holidays_0, df_holidays_1, df_holidays_2, df_holidays_3, df_next_day_0],
               ignore_index=True, sort=True)
```

```python
df.sort_values(by=['Date'], inplace=True, ascending=True)
df.head(n=10)
```
