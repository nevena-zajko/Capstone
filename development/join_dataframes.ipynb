```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns; sns.set(style="white", color_codes=True)
```

```python
# create dataframes for three datasets
df_incidents = pd.read_pickle("incident_data.pkl")
df_holidays = pd.read_pickle("holidays_data.pkl")
df_weather = pd.read_pickle("weather_data.pkl")
```

```python
# merge incident and weather datasets on Date and Time of Day
df = pd.merge(df_incidents, df_weather, on=['Date', 'TimeOfDay'])
```

```python
# rearrange columns
df = df[['Date', 'Month', 'DayOfWeek', 'TimeOfDay', 'Arrond',
         'Temp', 'RelHum', 'Pressure', 'WindSpd',
         'TotalPrecip', 'CumulPrecip', 'SnowOnGrnd', 'WindPwr', 'NumUnits']]
df.head()
```

```python
df_holidays.head()
```

```python
# merge holiday data
df = pd.merge(df, df_holidays, how='left', left_on=['Date', 'TimeOfDay'], right_on=['Date','TimeOfDay'])
# set 'Holiday' to 0 if not a holiday
df = df.fillna(0)
```

```python
# set datat types
df = df.astype({"Holiday": int})
df = df.astype({"NumUnits": int})
```

```python
df.head()
```

```python
df =  df.groupby(['Arrond', 'Month', 'DayOfWeek', 'TimeOfDay', 
             'Temp', 'RelHum', 'Pressure', 'WindSpd', 'TotalPrecip', 'CumulPrecip',
             'SnowOnGrnd', 'WindPwr', 'Holiday']).size().rename('count').reset_index()
```

```python
# write dataframe to file
df.to_pickle('incident_plus_weather.pkl')
```
